```python
# import library

import pandas as pd
```

```python
# load dataset

dataset=pd.read_csv("kidney_disease.csv")
dataset
```

```python
# Qualitative and Quantitative Data
# Categorical and Numerical data

qual=[]
quan=[]
for columnName in dataset.columns:
    if(dataset[columnName].dtypes=='O'):
        qual.append(columnName)
    else:
        quan.append(columnName)
```

```python
def varianceSD(dataset,quan):
    descriptive=pd.DataFrame(index=["Variance","SD"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Variance"]=dataset[columnName].var()
        descriptive[columnName]["SD"]=dataset[columnName].std()
        
    return descriptive
```

```python
varianceSD(dataset,quan)
```
