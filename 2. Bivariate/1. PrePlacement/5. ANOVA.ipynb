```python
# Qualitative and Quantitative Data
# Categorical and Numerical data

qual=[]
quan=[]
for columnName in dataset.columns:
    if(dataset[columnName].dtypes=='O'):
        qual.append(columnName)
    else:
        quan.append(columnName)
```

```python
qual
```

```python
quan
```

```python
# Variance Inflattion Factor

from statsmodels.stats.outliers_influence import variance_inflation_factor

def VIF(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    
    # shape[0] - row and shape[1]-column
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
VIF(dataset[quan])
```

```python
# salary - IVF < 5 - moderately correlated
# remaining column - IVF > 5 - highly correlated

# VIF value changes when compared with random variables
```

```python
VIF(dataset[["ssc_p","hsc_p","etest_p","salary"]])
```

```python
VIF(dataset[["degree_p","mba_p","salary"]])
```

```python
VIF(dataset[["etest_p","salary"]])
```

# T-test
# Paired and Unpaired t-test

```python
# Paired t-test
# Dependent Sample
# same group under different condition
# categorical - numerical


from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['salary']
male1 = dataset[dataset['gender']=='M']['etest_p']
ttest_rel(male, male1)
```

```python
# UnPaired t-test
# InDependent Sample
# different group under same condition
# categorical - numerical


from scipy.stats import ttest_ind
dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['salary']
female = dataset[dataset['gender']=='F']['salary']
#print(male)
ttest_ind(male, female)
```

```python
dataset[dataset['gender']=='F']["salary"]
```

```python
# Paired t-test
# Dependent Sample
# same group under different condition

from scipy.stats import ttest_rel
#dataset=dataset.dropna()
yes = dataset[dataset['workex']=='Yes']['salary']
yes1 = dataset[dataset['workex']=='Yes']['etest_p']
ttest_rel(yes, yes1)
```

```python
# UnPaired t-test
# InDependent Sample
# different group under same condition

from scipy.stats import ttest_ind
dataset=dataset.dropna()
yes = dataset[dataset['workex']=='Yes']['salary']
no = dataset[dataset['workex']=='No']['salary']
#print(male)
ttest_ind(yes, no)
```

# Hypothesis Testing
# Null or Alternate Hypothesis

```python
# Hypothesis Testing
```

```python
# Paired t-test
# Dependent Sample
# same group under different condition

from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['salary']
male1 = dataset[dataset['gender']=='M']['etest_p']
ttest_rel(male, male1)
```

```python
# p value <= 0.05 reject null hypothesis accept alternate hypothesis
```

```python
# For paired t-test,

# p value =1.78 
# p value > 0.05
# Accept Null Hypothesis
# Reject Alternate Hypotheis
```

```python
# UnPaired t-test
# InDependent Sample
# different group under same condition

from scipy.stats import ttest_ind
dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['salary']
female = dataset[dataset['gender']=='F']['salary']
#print(male)
ttest_ind(male, female)
```

```python
# For unpaired t-test

# p value =0.03 
# p value < 0.05
# Reject Null Hypothesis
# Accept Alternate Hypotheis
```
