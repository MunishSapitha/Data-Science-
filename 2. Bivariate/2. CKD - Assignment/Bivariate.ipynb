```python
# Qualitative and Quantitative Data
# Categorical and Numerical data

qual=[]
quan=[]
for columnName in dataset.columns:
    if(dataset[columnName].dtypes=='O'):
        qual.append(columnName)
    else:
        quan.append(columnName)
```

```python
# Variance Inflattion Factor

from statsmodels.stats.outliers_influence import variance_inflation_factor

def VIF(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    
    # shape[0] - row and shape[1]-column
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
VIF(dataset[quan])
```

# T-test
# Paired and Unpaired t-test

```python
# Paired t-test
# Dependent Sample
# same group under different condition
# categorical - numerical


from scipy.stats import ttest_rel
#dataset=dataset.dropna()
ckd = dataset[dataset['classification']=='ckd']['age']
ckd1 = dataset[dataset['classification']=='ckd']['hemo']
ttest_rel(ckd, ckd1)
```

```python
# UnPaired t-test
# InDependent Sample
# different group under same condition

from scipy.stats import ttest_ind
dataset=dataset.dropna()
ckd = dataset[dataset['classification']=='ckd']['age']
ckd1 = dataset[dataset['classification']=='notckd']['age']
#print(male)
ttest_ind(ckd, ckd1)
```
