```python
import pandas as pd
```

```python
dataset=pd.read_csv("kidney_disease.csv")
dataset
```

```python
def freqTable(columnName):
    freqTable=pd.DataFrame(columns=["Unique_Values","Frequency","Relative_F","Cumulative_RF"])
    freqTable["Unique_Values"]=dataset[columnName].value_counts().index
    freqTable["Frequency"]=dataset[columnName].value_counts().values
    freqTable["Relative_F"]=freqTable["Frequency"]/len(freqTable["Unique_Values"])
    freqTable["Cumulative_RF"]=freqTable["Relative_F"].cumsum()
    
    return freqTable
```
