# Data Preprocessing
# 1. Missing Values

```python
dataset.isnull().sum()
```

```python
# First consider numerical column
# 1. id - no null values
# 2. age - 9
# 3. bp - 12
# 4. sg - 47
# 5. al - 46
# 6. su - 49
# 7. bgr - 44
# 8. bu - 19
# 9. sc - 17
# 10.sod - 87
# 11. pot - 88
# 12. hemo - 52
```

```python
# age - mode()[0]

dataset["age"].fillna(dataset["age"].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# mean()

for i in ["bp","sg","al","su","bgr","bu","sc","sod","pot","hemo"]:
    dataset[i].fillna(dataset[i].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# mode() - categorical column - qual
 
# for i in qual:
#     dataset[i].fillna(dataset[i].mode(),inplace=True)
```

```python
for columnName in qual:
    print(columnName +" - "+ dataset[columnName].mode()[0])   
```

```python
# Replace categorical column with mode values

dataset['rbc'].fillna("normal",inplace=True)
dataset['pc'].fillna("normal",inplace=True)
dataset['pcc'].fillna("notpresent",inplace=True)
dataset['ba'].fillna("notpresent",inplace=True)
dataset['pcv'].fillna("41",inplace=True)
dataset['wc'].fillna("9800",inplace=True)
dataset['rc'].fillna("5.2",inplace=True)
dataset['wc'].fillna("9800",inplace=True)
dataset['htn'].fillna("no",inplace=True)
dataset['dm'].fillna("no",inplace=True)
dataset['cad'].fillna("no",inplace=True)
dataset['appet'].fillna("no",inplace=True)
dataset['pe'].fillna("no",inplace=True)
dataset['ane'].fillna("no",inplace=True)
```
